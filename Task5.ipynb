```python
import pandas as pd
import plotly.io as pio
```

```python
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output
from datetime import datetime
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
```

```python
# Load the datasets
apps_df = pd.read_csv('Play Store Data_1.csv')
reviews_df = pd.read_csv('User Reviews_1.csv')
```

```python
#Data Cleaning
apps_df = apps_df.dropna(subset=['Rating'])
for column in apps_df.columns :
    apps_df[column].fillna(apps_df[column].mode()[0],inplace=True)
apps_df.drop_duplicates(inplace=True)
apps_df=apps_df=apps_df[apps_df['Rating']<=5]
```

```python
#Convert the Installs columns to numeric by removing commas and +
apps_df['Installs']=apps_df['Installs'].str.replace(',','').str.replace('+','').astype(int)
```

```python
# Convert 'Reviews' to int
apps_df['Reviews'] = apps_df['Reviews'].astype(int)
```

```python
#Convert Price column to numeric after removing $
apps_df['Price']=apps_df['Price'].str.replace('$','').astype(float)
```

```python
apps_df['Size'] = apps_df['Size'].replace('Varies with device', '0')
```

```python
apps_df['Size'] = apps_df['Size'].str.replace('Varies with device', '0') 
```

```python
apps_df['Size'] = apps_df['Size'].str.replace('M', '')
```

```python
apps_df['Size'] = apps_df['Size'].str.replace('k', 'e-3')
```

```python
apps_df
```

```python
apps_df['Size'] = pd.to_numeric(apps_df['Size'], errors='coerce')
```

```python
apps_df['Size'].fillna(0, inplace=True)
```

```python
# Filter data as per task requirements
filtered_data = apps_df[
    (apps_df['Rating'] > 3.5) &
    (apps_df['Category'] == 'GAME') &
    (apps_df['Installs'] > 50000)
]
```

```python
# Create the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Play Store Apps Analysis"),
    dcc.Graph(id='bubble-chart'),
    dcc.Interval(id='interval-component', interval=1*1000, n_intervals=0)  # Refresh every second
])

# Callback to update the chart based on time
@app.callback(
    Output('bubble-chart', 'figure'),
    [Input('interval-component', 'n_intervals')]
)
def update_chart(n):
    # Get current hour
    current_hour = datetime.now().hour
    # Check if current time is between 12 PM and 4 PM
    if 12 <= current_hour < 16:
        fig = px.scatter(
            filtered_data,
            x='Size',
            y='Rating',
            size='Installs',
            color='Category',
            hover_name='App',
            title='App Size vs. Rating',
            labels={'Size': 'App Size (MB)', 'Rating': 'Average Rating'},
            size_max=60
        )
    else:
        fig = px.scatter(title='Data not available outside 12 PM - 4 PM')

    return fig

# Save the initial chart as an HTML dashboard
pio.write_html(update_chart(0), file='bubble_chart_dashboard.html', auto_open=True)

# Run the app on a different port, e.g., 8051
if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
